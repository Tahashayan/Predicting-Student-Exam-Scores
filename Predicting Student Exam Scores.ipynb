```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1,3,5,10])
y_train = np.array([20,30,50,100])
print(f"X_train is: {x_train}")
print(f"Y_train is: {y_train}")
```

```python
print(f"X_train.shape is: {x_train.shape}")
m = len(x_train)
print(f"The number of training examples is: {m}")
```

```python
for i in range(m):
    x_i = x_train[i]
    y_i = y_train[i]
    print(f"(x^{i},y^{i}) = ({x_i,y_i})")
```

```python
plt.scatter(x_train, y_train, marker = 'X', c = 'r')
plt.title("Student Exam Score")
plt.xlabel("Hours studied")
plt.ylabel("Exam scores")
plt.show()
```

```python
w = 9
b = 10
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_marks(x,w,b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
marks = compute_marks(x_train, w, b)
plt.plot(x_train, marks, c = 'b', label = "prediction")
plt.scatter(x_train, y_train, marker = 'X', c = 'r', label = 'Actual values')
plt.title("Student Exam score")
plt.xlabel("Hours studied")
plt.ylabel("Exam score")
plt.legend()
plt.show()
```

```python
def compute_score(x,y,w,b):
    m = x.shape[0]
    marks_score = 0
    for i in range(m):
        f_wb = w * x[i] + b
        score = (f_wb - y[i]) ** 2
        marks_score = marks_score + score
    total_score = (1 / (2 * m) * score)
    
    return total_score
```

```python
total_marks = compute_score(x_train, y_train, 9, 10)
print(total_marks)
```
